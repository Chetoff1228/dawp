```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
vector = (np.random.rand(1000) > 0.5).astype(int) * 2 - 1
vector[0] = 0
length = 0
flag = False


### мой тест

my_list = np.zeros(1000)
my_list[1:20] = 1
my_list[20:50] = -1
my_list[50:100] = -1
my_list[100:300] = -1
my_list[300:500] = 1
my_list[500:1000] = -1

my_list_cum = np.cumsum(my_list)
vector = my_list



## Task #2

for i in range(999):
   
    if vector[i] == vector[i+1]:
        length = length + 1
    else:
        length = 0
        
    if length == 9:
        flag = True
        first_index = i - 9
        
    if flag == True and length == 0:
        last_index = i+1
        break
        
## Task #3
length = 0
max_length = 0

for i in range(999):
   
    if vector[i] == vector[i+1]:
        length = length + 1
    else:
        if length > max_length:
            max_length = length
            max_last_index = i + 1
            max_first_index = i - 1 - length
        length = 0
```

```python
vis_vector  = np.cumsum(vector)
minimum = vis_vector.argmin(), vis_vector.min()
maximum = vis_vector.argmax(), vis_vector.max()
```

```python
plt.figure(figsize = (16, 10));
plt.title("Random Walk", fontsize = 20)
plt.plot(vis_vector);
plt.scatter(*minimum, c = "Red");
plt.scatter(*maximum, c = "Green");
plt.text(*minimum, f"Min point = {minimum[1]}", fontsize = 14);
plt.text(*maximum, f"Max point = {maximum[1]}", fontsize = 14);
plt.plot(np.arange(first_index, last_index), vis_vector[first_index:last_index], label = ">= 10 steps");
plt.plot(np.arange(max_first_index, max_last_index), vis_vector[max_first_index:max_last_index], ".", label = f"max steps = {max_length}", lw = 2);
plt.legend(fontsize = 16);
```
