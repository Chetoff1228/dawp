```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
ones_and_minus_ones = np.random.choice([-1,1],1000)
our_vec =list(ones_and_minus_ones)
our_vec.insert(0,0)
our_vec= np.array(our_vec)
our_vec = np.cumsum(our_vec)
our_vec
```

```python
np.min(our_vec)
```

```python
np.max(our_vec)
```

```python
np.argmin(our_vec)
```

```python
plt.figure(figsize=(15,8))
plt.plot(our_vec)
plt.axvline(np.argmin(our_vec) ,color ='r')
plt.axvline(np.argmax(our_vec) ,color ='green')
plt.legend(['блуждания','минимум','максимум'])
```
